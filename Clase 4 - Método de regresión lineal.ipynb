```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('Datasets/velocidad.csv', index_col = 0)
data.head(5)
```

```python
plt.scatter(data['t'],data['v'])
plt.xlabel('t')
plt.ylabel('v')
plt.show()
```

```python
x = ['t']
x = np.array(data[x])

y = ['v']
y = np.array(data[y])
```

```python
from sklearn.model_selection import train_test_split
#80% entrenamiento - 20%prueba - se generan cuatro arreglos.
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

### Regresión Lineal - Scikit-learn

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_,lr.coef_
```

$$ y = \theta x $$
$$ y = 3.02601651x + 3.89726834 $$

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, y_pred, color = 'red')
plt.show()
```

```python
x_new = np.array([[0.0],[2.0]])
```

```python
y_predict = lr.predict(x_new)
```

```python
plt.scatter(x,y, color = 'red')
plt.plot(x_new, y_predict, color = 'blue')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
